```python
from functools import partial 
def get_grade(position,grades):
    s_grades = str(grades)
    if len(s_grades) > 1:
        return int(s_grades[position])
    return 0
```

```python
train["grades_0"]  = train["grades"].apply(partial(get_grade,0))
train["grades_1"]  = train["grades"].apply(partial(get_grade,1))
train["grades_2"]  = train["grades"].apply(partial(get_grade,2))
train["grades_3"]  = train["grades"].apply(partial(get_grade,3))
train["grades_4"]  = train["grades"].apply(partial(get_grade,4))
```

```python
train.head(4)
```

```python
import re
def prepare_edited_new_headline(row):
    original_head_line = row.original
    edit = row.edit
    return re.sub(r'<.*/>', edit, original_head_line)
```

```python
train["edited_head_line"]= train.swifter.apply(prepare_edited_new_headline,axis=1)
dev["edited_head_line"]= dev.swifter.apply(prepare_edited_new_headline,axis=1)
```

```python
train.head(5)
```

```python
def get_original_headline(headline):
    return headline.replace("<","").replace("/>","")
```

```python
original_headlines_train = train["original"].swifter.apply(get_original_headline)
original_headlines_dev = dev["original"].swifter.apply(get_original_headline)
edited_headlines_train = train["edited_head_line"]
edited_headlines_dev = dev["edited_head_line"]
```

```python
all_original_headlines = original_headlines_train.append(original_headlines_dev)
all_edited_headlines = edited_headlines_train.append(edited_headlines_train)
```

```python
from nltk.corpus import stopwords 


from sklearn.feature_extraction.text import TfidfVectorizer
original_vectorizer = TfidfVectorizer(stop_words=stopwords.words("english"),ngram_range=(2,2))
original_vectorizer.fit(all_original_headlines.values.tolist())

from sklearn.feature_extraction.text import TfidfVectorizer
edited_vectorizer = TfidfVectorizer(stop_words=stopwords.words("english"),ngram_range=(2,2))
edited_vectorizer.fit(all_edited_headlines.values.tolist())

```

```python
train_features_original = original_vectorizer.transform(original_headlines_train)
train_features_edit = edited_vectorizer.transform(edited_headlines_train)

dev_features_original = original_vectorizer.transform(original_headlines_dev)
dev_features_edit = edited_vectorizer.transform(edited_headlines_dev)
```

```python
import numpy as np
train_features = np.hstack((train_features_original.todense(),train_features_edit.todense()))
dev_features = np.hstack((dev_features_original.todense(),dev_features_edit.todense()))
```

```python
train_features_df = pd.DataFrame(train_features,columns=original_vectorizer.get_feature_names() + edited_vectorizer.get_feature_names())
dev_features_df = pd.DataFrame(dev_features,columns=original_vectorizer.get_feature_names() + edited_vectorizer.get_feature_names())
```

```python
ntrain = train_features_df.shape[0]
ntest = dev_features_df.shape[0]
SEED = 0 # for reproducibility
NFOLDS = 4 # set folds for out-of-fold prediction
# kf = KFold(ntrain, n_folds= NFOLDS, random_state=SEED)

# Class to extend the Sklearn classifier
class SklearnHelper(object):
    def __init__(self, clf, seed=0, params=None):
        params['random_state'] = seed
        self.clf = clf(**params)

    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)

    def predict(self, x):
        return self.clf.predict(x)
    
    def fit(self,x,y):
        return self.clf.fit(x,y)
    
    def feature_importances(self,x,y):
        print(self.clf.fit(x,y).feature_importances_)
```

```python
def get_oof(x_train, train, x_test,grade):
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest ))
    test = train[grade].values
    
    skf = StratifiedKFold(n_splits=4)
    splits = skf.get_n_splits(train_features_df,train[grade].values)
    for i, (train_index, test_index) in enumerate(skf.split(train_features_df, test)):
        print("Doing for fold " + str(i))
        x_tr = x_train[train_index]
        y_tr = test[train_index]
        x_te = x_train[test_index]
        
        clf = LinearSVC(random_state=0,max_iter=10000,verbose=2)
        clf.fit(x_tr, y_tr)

        oof_train[test_index] = clf.predict(x_te)
        oof_test_skf[i, :] = clf.predict(x_test)

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
```
